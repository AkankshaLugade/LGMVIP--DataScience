```python
x=iris.sepallength
plt.hist(x)
plt.title("Histogram for sepal length")
plt.axvline(x.mean(),color='r',linestyle='dashed',linewidth=2)
plt.axvline(x.median(),color='g',linestyle='dashed',linewidth=2)
plt.axvline(x.min(),color='y',linestyle='dashed',linewidth=2)
plt.axvline(x.max(),color='y',linestyle='dashed',linewidth=2)
plt.xlabel("Sepal length")
plt.ylabel("Frequency")
plt.show()
```
